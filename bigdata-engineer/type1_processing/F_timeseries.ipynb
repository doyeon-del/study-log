# F. 시계열


개념 핵심 요약: 

> 판다스에서 날짜·시간 데이터는 단순 문자열(object)이 아닌 datetime64 타입이어야 연도, 월, 일, 시간 단위로 자유롭게 분리 및 연산이 가능하다. 이를 위해 pd.to_datetime() 함수를 필수로 사용한다.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# 시계열 (datetime)



```python
df_time = pd.DataFrame({
    'Date': ['2026-01-15', '2026-02-28', '2026-06-02']
})


## 문자열 타입을 datetime 타입으로 변환하기!!
df_time['Date'] = pd.to_datetime(df_time['Date'])

## dt 접근자활용해서 연, 월, 일 추출하기
df_time['Year'] = df_time['Date'].dt.year
df_time['Month'] = df_time['Date'].dt.month
df_time['Day'] = df_time['Date'].dt.day

print(df_time)
```
