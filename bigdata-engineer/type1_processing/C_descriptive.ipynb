## 표준편차


**풀이 메모:** df.corr(numeric_only = True), .std()

```python
df = sns.load_dataset('titanic')


# 주의 - 수치형 변수에 대해서만 계산 가능함! - 필터링 필요함.

df_num = df.corr(numeric_only = True)

df_num.std()
```
