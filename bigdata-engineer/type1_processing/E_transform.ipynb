## Binning Data
- 연속형 변수(예: 나이, 운임)를 '청소년', '청년', '장년'처럼 범주형 그룹으로 나누는 작업

  - age를 3개 구간으로 나눠보기 --> pandas의 cut을 활용하여 집단별로 나누기!!

```python

bins = [0, 19, 59, 100]
labels = ['Child/Teen', 'Adult', 'Senior']

df['age_group'] = pd.cut(df['age'], bins = bins, labels = labels)
df['age_group'].value_counts()
```
