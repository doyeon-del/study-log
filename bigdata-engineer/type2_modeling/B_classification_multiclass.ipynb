## T2-4 펭귄 종 분류 (macro-F1)

**문제 정의**

신체 측정값으로 펭귄 종(`species`, 3종)을 예측하라.
* label: `species` (Adelie/Chinstrap/Gentoo)
* 제출: 예측 **클래스**를 `pred` 컬럼으로 `result.csv` 저장
* 평가: **macro F1** (`f1_score(..., average='macro')`)
* 주의: 범주형(`island`,`sex`)·결측치 처리 필요

---

**문제 요약(본인 말로):** 

**풀이 메모:** 

```python
df = sns.load_dataset('penguins').dropna(subset=['species']).reset_index(drop=True)
train, test = train_test_split(df, test_size=0.3, random_state=42, stratify=df['species'])
y_test = test['species']                                    # 채점용
test  = test.drop(columns='species').reset_index(drop=True)
train = train.reset_index(drop=True)
train.head()
```
