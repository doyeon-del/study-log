```python
import warnings; warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
```

## T2-1 생존 예측 (ROC-AUC, 확률 제출)

**문제 정의**

타이타닉 탑승객 정보로 생존 여부(`survived`)를 예측하라.
* label: `survived` (0/1)
* 제출: 생존할 **확률**을 `pred` 컬럼 1개로 `result.csv` 저장
* 평가: **ROC-AUC**

---

**문제 요약(본인 말로):** 

**풀이 메모:** 

```python
df = sns.load_dataset('titanic')[['survived','pclass','sex','age','sibsp','parch','fare','embarked']]
train, test = train_test_split(df, test_size=0.3, random_state=42, stratify=df['survived'])
y_test = test['survived']                                   # 채점용
test  = test.drop(columns='survived').reset_index(drop=True)
train = train.reset_index(drop=True)
train.head()
```

## T2-2 유방암 진단 (F1, 클래스 제출)

**문제 정의**

세포 측정값으로 양성/악성 여부(`target`)를 예측하라.

**문제 요약(본인 말로):** 

**풀이 메모:** 

```python
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer(as_frame=True).frame
train, test = train_test_split(df, test_size=0.3, random_state=42, stratify=df['target'])
y_test = test['target']                                     # 채점용
test  = test.drop(columns='target').reset_index(drop=True)
train = train.reset_index(drop=True)
train.shape, test.shape
```

## T2-3 불균형 채무불이행 (F1, 성능개선)

**문제 정의**

신청자 데이터로 채무 불이행 여부를 예측하라(양성 비율이 낮은 불균형 데이터).
* label: `target` (0: 이행, 1: 불이행)
* 제출: 예측 **클래스**를 `pred` 컬럼으로 `result.csv` 저장
* 평가: **F1-score**
* 힌트: `class_weight='balanced'`, 임계값 조정, `scale_pos_weight` 등으로 개선 시도

---

**문제 요약(본인 말로):** 

**풀이 메모:** 

```python
from sklearn.datasets import make_classification
X, yv = make_classification(n_samples=2000, n_features=12, n_informative=6,
                            weights=[0.9, 0.1], random_state=42)
df = pd.DataFrame(X, columns=[f'f{i}' for i in range(12)]); df['target'] = yv
train, test = train_test_split(df, test_size=0.3, random_state=42, stratify=df['target'])
y_test = test['target']                                     # 채점용
test  = test.drop(columns='target').reset_index(drop=True)
train = train.reset_index(drop=True)
train['target'].value_counts(normalize=True)
```
